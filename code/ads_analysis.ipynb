```python
us_pres_votes = pd.read_csv("../data/us_president_2016_votes_2.csv")
```

```python
us_pres_votes
```

```python
state_wise_votes = us_pres_votes.groupby(["state","candidate"], as_index=False).aggregate({"votes":"sum"})
```

```python
state_wise_votes_HT = state_wise_votes[(state_wise_votes["candidate"] == "Donald Trump") | 
                                      (state_wise_votes["candidate"] == "Hillary Clinton")]
```

```python
state_wise_votes_T = np.array(state_wise_votes_HT[state_wise_votes["candidate"] == "Donald Trump"]["votes"])
```

```python
state_wise_votes_T
```

```python
state_wise_votes_HT["state"].unique()
```

```python
state_wise_votes_H = np.array(state_wise_votes_HT[state_wise_votes["candidate"] == "Hillary Clinton"]["votes"])
```

```python
state_wise_votes_H
```

```python
state_wise_votes_T - state_wise_votes_H
```

```python
state_wise_votes_T_pd = state_wise_votes[state_wise_votes["candidate"] == "Donald Trump"].copy()
```

```python
state_wise_votes_T_pd["diff_with_Hillary"] = state_wise_votes_T - state_wise_votes_H
```

```python
state_wise_votes_T_pd.to_csv("../data/state_wise_vote_diff.csv", index=False)
```

```python
county_wise_votes = us_pres_votes.groupby(["county_name","candidate"], as_index=False).aggregate({"votes":"sum"})
```

```python
county_wise_votes
```

```python
county_wise_votes_T = np.array(county_wise_votes[county_wise_votes["candidate"] == "Donald Trump"]["votes"])
```

```python
county_wise_votes_H = np.array(county_wise_votes[county_wise_votes["candidate"] == "Hillary Clinton"]["votes"])
```

```python
county_wise_votes_diff = county_wise_votes_T - county_wise_votes_H
```

```python
county_wise_votes_T_pd = county_wise_votes[county_wise_votes["candidate"] == "Donald Trump"].copy()
```

```python
county_wise_votes_T_pd["diff_with_Hillary"] = county_wise_votes_diff
```

```python
florida_counties = us_pres_votes[us_pres_votes["state"] == "Florida"]["county_name"].unique()
```

```python
florida_counties
```

```python
florida_results = county_wise_votes_T_pd[county_wise_votes["county_name"].isin(florida_counties)].copy()
```

```python
florida_results[florida_results["diff_with_Hillary"] > 0].sort_values("diff_with_Hillary",ascending=False).count()
```
