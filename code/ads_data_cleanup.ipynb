```python
# import libraries

import pandas as pd
```

```python
# read the file
tv_ads = pd.read_csv("../data/PoliticalTVAds_2015_export.csv")
```

```python
tv_ads.head()
```

```python
# extract the year and date when the ad was aired

tv_ads["year_aired"] = pd.DatetimeIndex(tv_ads["start_time"]).year

tv_ads["month_aired"] = pd.DatetimeIndex(tv_ads["start_time"]).month
```

```python
tv_ads.head()
```

```python
# separate the cities and states information from the location column

tv_ads["location_split"] = tv_ads["location"].str.split('/')

tv_ads_exploded_1 = tv_ads.explode("location_split")

tv_ads_exploded_1.reset_index(inplace=True)

states = tv_ads_exploded_1["location_split"].str.split(",", expand=True)

tv_ads_exploded_1["cities_aired"] = states[0]

tv_ads_exploded_1["states_aired"] = states[1]
```

```python
tv_ads_exploded_1
```

```python
# split the individual cities from the cities_aired column into individual rows

tv_ads_exploded_1["cities_aired_split"] = tv_ads_exploded_1["cities_aired"].str.split("-")

tv_ads_exploded_2 = tv_ads_exploded_1.explode("cities_aired_split")

tv_ads_exploded_2.reset_index(inplace=True)

tv_ads_clean = tv_ads_exploded_2.drop(["level_0", "index", "location_split", "cities_aired"], axis=1).copy()
```

```python
tv_ads_clean
```

```python
# correct the format of the rows labeled Iowa in the states_aired column
tv_ads_clean["states_aired"] = tv_ads_clean["states_aired"].str.replace(pat="Iowa", repl="IA")
```

```python
tv_ads_clean["states_aired"].unique()
```

```python
# correct the format of the rows labeled Ohio in the states_aired column
tv_ads_clean["states_aired"] = tv_ads_clean["states_aired"].str.replace(pat="Ohio", repl="OH")
```

```python
tv_ads_clean["cities_aired"] = tv_ads_clean["cities_aired_split"]
```

```python
tv_ads_clean.drop(["cities_aired_split"], axis=1, inplace=True)
```

```python
tv_ads_clean
```

```python
# collect county information for all the us cities
county_info = pd.read_csv("../data/us_cities.csv")
```

```python
tv_ads_clean["cities_aired"].unique()
```

```python
tv_ads_clean["cities_aired"] = tv_ads_clean["cities_aired"].str.replace("New York City", "New York")
```

```python
county_info["CITY"] = county_info["CITY"].astype(str)
```

```python
county_info["STATE_CODE"] = county_info["STATE_CODE"].astype(str)
```

```python
tv_ads_clean["cities_aired"] = tv_ads_clean["cities_aired"].astype(str)
```

```python
tv_ads_clean["states_aired"] = tv_ads_clean["states_aired"].astype(str)
```

```python
# cleaning leading white spaces in the states_aired column
tv_ads_clean["states_aired"] = tv_ads_clean["states_aired"].str.lstrip()
```

```python
# merge the tv ads data with the counties table to find the matching counties
# for each of the cities where the ads were aired.

tv_ads_clean = tv_ads_clean.merge(county_info, how="left", left_on=["cities_aired", "states_aired"], 
                  right_on=["CITY", "STATE_CODE"])
```

```python
tv_ads_clean.columns
```

```python
# filter in the ads for the presidential collection
tv_ads_clean_PRES = tv_ads_clean[tv_ads_clean["race"] == "PRES"].copy()
```

```python
# filter in the ads aired in 2016
tv_ads_clean_PRES_2016 = tv_ads_clean_PRES[tv_ads_clean_PRES["year_aired"] == 2016].copy()
```

```python
tv_ads_clean_PRES_2016
```

```python
# write the records into a csv file
tv_ads_clean_PRES_2016.to_csv("../data/tv_ads_cleaned_2016.csv")
```

```python
# cleaning the county name column for the trailing repetitions in each row
us_election_2016["county_name"] = us_election_2016["county_name"].str.rstrip(" County")
```
